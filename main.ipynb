```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('movies.csv')
movies['title'] = movies['title'].str[:-7]
movies['genres'] = movies['genres'].str.replace('|', ' ')

ratings = pd.read_csv('ratings.csv')
movies
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(movies['genres'].values)
feature_names = vectorizer.get_feature_names_out()
movie_embeddings = pd.DataFrame(X.toarray(), columns=feature_names)
movie_embeddings['vector'] = movie_embeddings.values.tolist()
movies['genres'] = movie_embeddings['vector']
movies.to_csv('temp.csv')
```

```python
def get_sim(movieid):
    movie_to_search = movies[['movieId', 'genres']]
    movie_to_search = movie_to_search[movie_to_search.movieId != movieid]
    movie_to_search['similarity'] = movie_to_search['genres'].apply(lambda x: cosine_similarity(np.array(x).reshape(1, -1), np.array(movies[movies['movieId'] == movieid]['genres'].values[0]).reshape(1, -1)))
    return movie_to_search
sim = get_sim(2)
sim
```
